**Enunciado 1:** dado el archivo de registros de jugadas (eventos) de MemPy, mostrar un gráfico con el porcentaje de partidas que se juegan los fines de semana (sábados y domingos).

```
import pandas as pd
import os
from datetime import datetime
import time
from matplotlib import pyplot as plt
import numpy as np

dias_semana = [0, 1, 2, 3, 4]
dias_finde = [5, 6]

cant_sem = 0
cant_finde = 0

df_juego = pd.read_csv(os.path.join('datos_de_prueba.csv'))
df_partidas = df_juego[df_juego['Nombre de evento']=='inicio_partida'][['Partida','Tiempo']]

for x in df_partidas.index:
  if datetime.fromtimestamp(df_partidas.loc[x, 'Tiempo']).weekday() in dias_semana:
    cant_sem += 1
  else:
    cant_finde += 1

data = np.array([cant_sem, cant_finde])
etiquetas = ["Partidas días de semana", "Partidas fines de semana"]
explode = (0.1, 0)  # la separación entre las partes.

plt.pie(data, explode=explode, labels=etiquetas, autopct='%1.1f%%',
        shadow=True, startangle=90, labeldistance= 1.1) # autopct: cantidad de decimales.
plt.axis('equal')
plt.legend(etiquetas)  # el cuadrito que explica los colores.

plt.show()
```
